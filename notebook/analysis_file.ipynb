## **BUSINESS UNDERSTANDING**

### **Introduction**

Pertumbuhan populasi dunia merupakan isu strategis yang memengaruhi aspek ekonomi, sosial, dan lingkungan. Jumlah penduduk yang besar dapat menjadi potensi sumber daya manusia, namun juga menimbulkan tantangan dalam penyediaan pangan, energi, pendidikan, dan kesehatan. Dinamika populasi tidak hanya ditentukan oleh angka kelahiran dan kematian, tetapi juga migrasi, urbanisasi, serta distribusi usia. Negara dengan angka kelahiran tinggi menghadapi tekanan pembangunan, sedangkan negara dengan populasi menua berisiko mengalami penurunan produktivitas. Oleh karena itu, analisis data populasi global diperlukan untuk memahami kondisi demografi dan mendukung kebijakan pembangunan berkelanjutan.

### **Problem Statement**

Berdasarkan latar belakang tersebut, maka permasalahan yang ingin dikaji adalah:

1. Bagaimana distribusi jumlah populasi tiap negara serta persentasenya terhadap total populasi dunia?

2. Negara mana saja yang memiliki tingkat pertumbuhan penduduk tertinggi dan terendah, serta bagaimana gambaran rata-rata angka kelahiran (fertility rate) dunia?

3. Bagaimana kepadatan penduduk setiap negara jika dilihat dari keterkaitannya dengan luas wilayah, jumlah populasi, dan tingkat urbanisasi?

4. Negara mana yang memiliki tingkat migrasi tertinggi.

5. Berapakah rata-rata usia penduduk dunia (median age)?

### **Objectives**

1. Menganalisis distribusi populasi dunia dengan menghitung jumlah populasi setiap negara serta kontribusinya dalam persentase terhadap total populasi global.

2. Mengidentifikasi negara dengan pertumbuhan populasi tertinggi, serta menghitung rata-rata angka kelahiran (fertility rate) untuk mendapatkan gambaran kondisi demografi global.

3. Mengevaluasi kepadatan penduduk setiap negara dengan mengaitkannya pada faktor luas wilayah, jumlah populasi, serta tingkat urbanisasi, sehingga diperoleh pemahaman mengenai persebaran penduduk yang lebih komprehensif.

4. Menentukan negara dengan tingkat migrasi tertinggi.

5. Menghitung rata-rata usia penduduk dunia (median age).

## **DATA UNDERSTANDING**

### **Deskrpisi Variabel Dataset:**

Dataset yang digunakan merupakan hasil scraping dari website worldometers:https://www.worldometers.info/world-population/population-by-country/

1. **`Population`** : Jumlah total penduduk suatu negara pada tahun tertentu.

2. **`Yearly Change (%)`** : Persentase pertumbuhan penduduk per tahun dibandingkan tahun sebelumnya.

3. **`Net Change`** : Selisih jumlah penduduk yang bertambah setiap tahun (dalam angka, bukan persen).

4. **`Density (P/Km²)`** : Kepadatan penduduk, dihitung sebagai jumlah penduduk per kilometer persegi (orang per km²).

5. **`Land Area (Km²)`** : Luas daratan suatu negara dalam kilometer persegi (tidak termasuk perairan).

6. **`Migrants (net)`** : Angka migrasi bersih (imigran masuk dikurangi emigran keluar).
        Positif → lebih banyak orang masuk, Negatif → lebih banyak orang keluar.

7. **`Fertility Rate`** : Rata-rata jumlah anak yang dilahirkan seorang perempuan sepanjang hidupnya.

8. **`Median Age`** : Usia tengah penduduk (setengah lebih muda dari angka ini, setengah lebih tua).

9. **`Urban Population (%)`** : Persentase penduduk yang tinggal di wilayah perkotaan.

10. **`World Share (%)`** : Persentase penduduk suatu negara dibandingkan total populasi dunia.

## **DATA CLEANSING**

```python
# Copy dataset untuk dilakukan cleaning
data = df.copy()
```

### **Handling Missing Value**

```python
# Pengecekan data dengan urban population null
data[data['Urban Population (%)'].isnull()]
```

Berdasarkan hasil analisis sebelumnya diketahui bahwa Kolom Urban Population yang memiliki nilai Null merupakan negara kecil yang hanya memiliki 1 kota sehingga seluruh penduduknya dianggap sebagai penduduk urban sehingga nilainya menjadi 100%

```python
# Mengubah data Null menjadi 100% pada kolom Urban Population
data["Urban Population (%)"] = data["Urban Population (%)"].fillna('100%')

# Melakukan pengecekan data
data.isnull().sum()
```

Setelah dilakukan pembersihan data dengan nilai null maka sudah tidak ada lagi data yang bernilai null dikarenakan semua data telah diubah menjadi 100%.

### **Data Transformation**

Pada data tersebut hanya data pada kolom Fernility Rate dan Median Age saja yang merupakan data number sedangkan kolom lainnya bertipe object, sehingga perlu diubah tipe datanya menjadi numeric sehingga dapat digunakan pada proces analisis selanjutnya.

```python
# fungsi untuk membersihkan angka
def clean_number(x):
    x = str(x).replace(",", "")   # hapus koma ribuan
    x = x.replace("%", "")        # hapus persen
    x = x.replace("−", "-")       # ubah minus aneh ke minus normal
    return x.strip()

# copy biar data asli tetap ada
df_clean = data.copy()

# proses semua kolom kecuali 'Country'
for col in df_clean.columns:
    if col != "Country":
        df_clean[col] = df_clean[col].apply(clean_number)               # bersihkan
        df_clean[col] = pd.to_numeric(df_clean[col], errors="coerce")   # ubah ke angka

# tampilkan hasil tipe data
print('Tipe data setelah dilakukan perubahan:\n')
print(df_clean.dtypes)

# contoh hasil clean
df_clean.head()
```

```python
# Mengubah nama kolom sehingga lebih jelas
df_clean = df_clean.rename(columns={"Yearly Change": "Yearly Change (%)"})

df_clean.head()
```

### **Save data clean**

```python
# Menyimpan data yang sudah dilakukan cleaning

df_clean.to_csv('../data/world_population_data_clean.csv', index=False)
```

### **Distribusi Populasi Dunia per Negara**

**Top 5 Negara dengan jumlah populasi terbanyak di dunia**

```python
# Fungsi format angka
def format_number(n):
    if n >= 1_000_000_000:     # miliar
        return f"{n/1_000_000_000:,.1f} M"
    elif n >= 1_000_000:       # juta
        return f"{n/1_000_000:,.0f} Juta"
    else:
        return str(n)

# Ambil 10 negara dengan populasi terbesar
data_population = df_clean.nlargest(5, "Population").reset_index()

plt.figure(figsize=(12, 8))
bars = plt.bar(data_population['Country'], data_population['Population'], color="skyblue")

plt.title("Negara dengan Jumlah Populasi Terbesar", fontsize=14)
plt.xlabel("Negara")
plt.ylabel("Population (dalam Juta)")

# Rotasi label biar tidak bertumpuk
plt.xticks(rotation=45)

# Ubah sumbu Y jadi jutaan (angka bulat)
labels, locs = plt.yticks()
plt.yticks(labels, (labels/1_000_000).astype(int))

# Tambahkan label di atas bar dengan format Billion/Million
for bar in bars:
    yval = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width()/2,   # posisi X tengah bar
        yval,                              # posisi Y di atas bar
        format_number(yval),               # format pakai fungsi
        ha='center', va='bottom', fontsize=9, color="blue"
    )

plt.show()
```

**Insight:**

Grafik batang menunjukkan dominasi India dan China sebagai dua negara dengan jumlah penduduk terbanyak di dunia, masing-masing lebih dari 1,4 miliar jiwa. Kesenjangan jumlah penduduk dengan negara lain sangat besar, di mana Amerika Serikat, Indonesia, dan Pakistan berada jauh di bawah 400 juta. Hal ini menegaskan betapa pentingnya India dan China dalam peta demografi global.

**Persentase jumlah population dengan dunia**

```python
# Hitung total World Share untuk sisanya
others_value = df_clean["World Share (%)"].sum() - data_population["World Share (%)"].sum()

# Buat data untuk pie chart
sizes = list(data_population["World Share (%)"]) + [others_value]
labels = list(data_population["Country"]) + ["Others"]

plt.figure(figsize=(8,8))
plt.pie(sizes, labels=labels, autopct="%1.1f%%", startangle=140)
plt.title("Top 5 Negara Penyumbang Populasi Dunia")
plt.show()
```

**Insight:**

Berdasarkan pie chart tersebut, terlihat bahwa India (17,8%) dan China (17,2%) menjadi negara dengan kontribusi populasi terbesar di dunia, bahkan jika digabung keduanya menyumbang lebih dari sepertiga total populasi global. Sementara itu, Amerika Serikat (4,2%), Indonesia (3,5%), dan Pakistan (3,1%) masuk dalam lima besar negara terpadat, namun kontribusinya masih jauh lebih kecil dibanding India dan China. Menariknya, lebih dari separuh penduduk dunia (54,2%) tersebar di luar lima negara tersebut, yang menunjukkan bahwa meskipun ada dua negara dengan dominasi populasi yang signifikan, distribusi penduduk dunia tetap cukup merata di banyak negara lain

### **Negara dengan Tingkat Migrasi Tertinggi**

```python
def format_number(n):
    if n >= 1_000_000:       # Juta
        return f"{n/1_000_000:.2f} Juta"
    elif n >= 1_000:           # Ribu
        return f"{n/1_000:.0f} Ribu"
    else:
        return str(n)


# Ambil 5 negara dengan migrasi masuk terbanyak (positif)
top5_migrants = df_clean.nlargest(5, "Migrants (net)")

plt.figure(figsize=(10,6))
bars = plt.barh(top5_migrants["Country"], top5_migrants["Migrants (net)"], color="lightsalmon")

plt.title("5 Negara dengan Migrasi Masuk Terbanyak (+)")
plt.xlabel("Migrants (net) (in Millions)")
plt.ylabel("Negara")

# Ubah xticks ke jutaan secara manual
labels, locs = plt.xticks()  
plt.xticks(labels, (labels/100000).astype(int))

# Tambahkan angka di ujung bar
for bar, value in zip(bars, top5_migrants["Migrants (net)"]):
    plt.text(value,
             bar.get_y() + bar.get_height()/2,
             format_number(value),   # gunakan format_number
             ha="left", va="center", fontsize=9, color="black")

plt.show()
```

**Insight:**

Hal ini mencerminkan daya tarik negara-negara tersebut, baik karena faktor ekonomi, politik, maupun keamanan. Negara-negara Eropa Barat seperti Inggris dan Kanada juga menjadi destinasi populer meskipun dalam jumlah lebih kecil.

### **Variasi Median Age seluruh negara didunia**

```python
plt.figure(figsize=(10, 5))
ax = sns.boxplot(x=df_clean["Median Age"], color="lightgreen")

# Judul & label
plt.title("Variasi Median Age Penduduk Dunia", fontsize=14, weight="bold")
plt.xlabel("Median Age (Tahun)", fontsize=12)

# Tambahkan garis rata-rata
mean_val = df_clean["Median Age"].mean()
plt.axvline(mean_val, color="red", linestyle="--", label=f"Mean = {mean_val:.1f}")

# Tambahkan garis median
median_val = df_clean["Median Age"].median()
plt.axvline(median_val, color="blue", linestyle="-.", label=f"Median = {median_val:.1f}")

# Nilai min & max
min_val = df_clean["Median Age"].min()
max_val = df_clean["Median Age"].max()

# Cari negara termuda & tertua
country_min = df_clean.loc[df_clean["Median Age"].idxmin(), "Country"]
country_max = df_clean.loc[df_clean["Median Age"].idxmax(), "Country"]

# Tambahkan anotasi min & max
plt.text(min_val, 0.05, f"Min: {min_val} ({country_min})", 
         ha="left", va="bottom", fontsize=9, color="black")
plt.text(max_val, 0.05, f"Max: {max_val} ({country_max})", 
         ha="right", va="bottom", fontsize=9, color="black")

plt.legend()
plt.show()
```

**Insight:**

Boxplot memperlihatkan bahwa usia median penduduk dunia cukup seimbang, dengan rata-rata sekitar 31–33 tahun. Negara dengan penduduk termuda adalah Republik Afrika Tengah (14,5 tahun), sementara yang tertua adalah Holy See (57,4 tahun). Perbedaan ini menyoroti kontras antara negara dengan populasi muda yang masih berkembang dengan negara yang menghadapi tantangan penuaan populasi.
